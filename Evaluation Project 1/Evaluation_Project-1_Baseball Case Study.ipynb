# About The Game : 'Baseball' 

Baseball is a bat and ball game played between two oppossing teams. Game consists of nine players each in both the teams and take turns batting and fielding. The game proceeds when a player on the fielding team, called the pitcher throws a ball towards a player of batting team (tries to hit with a bat).

The first objective of the team is to have a player reach first base safely. The player who reaches first abse safely without being called 'out' can attempt to advance to subsequent bases as a runner, either immediately or during teammates'turns batting. The fielding teams trys to prevent runs by getting batters or runners 'out', which forces them out of the field of play. The batting team's turn to bat is over once the fielding team records three outs. One turn batting for each team constitutes an inning. A game is usually composed of nine innings, and the team with the greater no. of runs at the end of the game wins. if the scores are tied at the end of nine innings, extra innings are usually played. Baseball has no game clock, although most games end in the ninth inning.

First played during mid 18th century in England. By the late 19th century, baseball was widely recognized as the national sport of the United States. In the US and Canada, professional Major League Baseball (MLB) teams are divided into the National League (NL) and American League(AL). The World Baseball Classic, organized by the World Baseball Softball Confedereation.

### Some Techncial Game Terms:

1) Home run (HR): It is scored when the ball is hit in such a way that the batter is able to circle the bases and reach home safely 
             in one play without any errors being commited by the defensive team in the process. In the modern day baseball,                the feat is typically achieved by hitting the ball over the outfield fence between the foul ploes (or making                    contact with either foul pole) without first touching the ground, resulting in an automatic home run. When a home              run is scored, the batter is also credited with a hit and a run scored.
             
2) At Bat (AB): An at-bat is a specific type of plate appearance in which the batter puts the ball in play intending to get on base. It is solely use to calculate batting average.

3) A Run Batted In (RBI): It is a statistic in baseball that credits a batter for making a play that allows a run to be scored (without an error is made on the play). For example, if the batter bats a base hit which allows a teammate on a higher base to reach home and so score a run, then the batter gets credited with a RBI.

4) Batting Average (BA): Playres hits (H) divided by his total at-bats (AB) i.e.(H/AB), is called batting average in baseball.

5) Runs(scored)-R: Number of times a player crosses home plate.

6) Hits (H): Reaching base beacause of batted, fair ball without error by the defence.

7) Doubles (2B): Hits on which the batter reaches second base safely without the contribution of fielding error.

8) Triples (3B): Hits on which the batter reaches third base safely without the contribution of fielding error.

9) Walks (Base on balls-BB): Times pitching four balls, allowing the batter to take first base.

10) Strike-outs (SO): Number of batters who recieved strike three (given out by the umpire).

11) Stolen Bases (SB): Number of bases advanced by the runner while the ball is in the possession of the defence.

12) Runs Allowed (RA): Total number of runs a pitcher allowed.

13) Earned Runs (ER): Number of runs that did not occur as a result of errors or passed balls.

14) Earned Run Average (ERA): It is determined by dividing the number of earned runs allowed by the number of innings pitch and multiplying by nine.

15) Shutouts (SHO): Number of complete games pitched with no runs allowed.

16) Saves (SV): Number of games where the pitcher enteres a game led by the pitcher's team, finishes the game without surrendering the lead, is not the winning pitcher, and either a) the lead was three runs or fewer when the pitcher entered the game; b) the potential tying run was on base, at bat, or on deck; or c) the pitcher pitched three or more innings.

17) Complete Games (CG): Number of games where player was the only pitcher for their team.

18) Errors (E): Number of times a fielder fails to make a play he should have made with common effort, and the offense (batter team) benefits as a result.

# Train Test Split and Finding Best Random State:

```python
import sklearn
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
from sklearn.metrics import mean_squared_error,mean_absolute_error
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')
```

```python
for i in range(0,1000):
    x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=i,test_size=0.20)
    lr.fit(x_train,y_train)
    pred_train=lr.predict(x_train)
    pred_test=lr.predict(x_test)
    if round(r2_score(y_train,pred_train)*100,1)==round(r2_score(y_test,pred_test)*100,1):
        print("At random state",i,"The model performs very well")
        print("At random_state:-",i)
        print("Training r2_score is:-",r2_score(y_train,pred_train)*100)
        print("Testing r2_score is:-",r2_score(y_test,pred_test)*100)    
```

Hence, we got model accuracy 93% in both training and testing at random_state=106. 
So, Selecting random_state=106.

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=106)
```

```python
lr.fit(x_train,y_train)
```

```python
pred_test=lr.predict(x_test)
```

```python
print('Mean Absolute Error:', mean_absolute_error(y_test,pred_test))
print('Mean Squared Error:', mean_squared_error(y_test,pred_test))
print('Root Mean Squared Error:', np.sqrt(mean_squared_error(y_test,pred_test)))
print('Model Accuracy:', r2_score(y_test,pred_test))
```

## Cross Validation Score for Linear Regression:

```python
pred_lr=lr.predict(x_test)
from sklearn.model_selection import cross_val_score
lss=r2_score(y_test,pred_lr)
for j in range(2,10):
    lsscore=cross_val_score(lr,x,y,cv=j)
    lsc=lsscore.mean()
    print("At cv:",j)
    print("Cross validation score is:",lsc*100)
    print("Model accuracy is:",lss*100)
    print("\n")
```

CV=8 is better for Linear Regression amongst all, althogh there is a vast difference between cv mean score and Model accuracy. Hence we will use other algorithms.
